```python
import matplotlib.pyplot as plt

# D liu v cc khon khu tr
deduction_categories = [
    "Federal Tax (25%)",
    "State Tax (5%)",
    "FICA (6.2%)",
    "Medicare (1.45%)",
    "Group Insurance ($600)",
    "Profit Sharing (5%)"
]
deduction_totals = [20000, 4000, 5000, 1200, 4800, 5000]  # Tng s tin cho mi loi khu tr

# To biu  dng ct
plt.figure(figsize=(10, 6))
plt.bar(deduction_categories, deduction_totals, color="orange")
plt.xlabel("Deduction Categories", fontsize=12)
plt.ylabel("Total Amount ($)", fontsize=12)
plt.title("Deductions Breakdown for Executives (Task 1-9)", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Hin th biu 
plt.show()
```
